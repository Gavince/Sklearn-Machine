### 3.描述性统计和异常值

#### 异常值的分析思路
0. 如果你发现了异常值，首先你要观察，这个异常值出现的频率　　
1. 如果异常值只**出现了一次**，多半是输入错误，直接把异常值删除　　
2. 如果异常值**出现了多次**，去跟业务人员沟通，可能这是某种特殊表示，如果是人为造成的错误，异常值留着是没有的，只要数据量不是太大，都可以删除　　
3. 如果异常值占到你总数据量的10%以上了，不能轻易删除。可以考虑把异常值替换成非异常但是非干扰的项，比如说用0来进行替换，或者把异常当缺失值，用均值或者众数来进行替换

### 7.建立模型

```python
from time import time
import datetime
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score, roc_curve, recall_score
```

```python
## 建模型
# 标签二维数据必须转化为一维度
Ytrain.shape
Ytrain = Ytrain.iloc[:, 0].ravel()
Ytest = Ytest.iloc[:, 0].ravel()
```

```python
t0 = time()
for kernel in ["linear", "rbf", "sigmoid", "poly"]:
    svc = SVC(kernel=kernel
              , degree=1
              , gamma="auto"
              , cache_size=5000).fit(Xtrain, Ytrain)
    result = svc.predict(Xtest)
    # recall召回率　捕捉少数类的能力
    score = svc.score(Xtest, Ytest)
    recall = recall_score(Ytest, result)
    auc = roc_auc_score(Ytest, svc.decision_function(Xtest))
    print("The kernle is {}, score : {}, recall : {}, auc : {}".format(kernel, score, recall, auc))
    print(datetime.datetime.fromtimestamp(time() - t0).strftime("%M:%S:%f"))
```

#### 7.1 模型调参

#### 调参方向
我们可以有不同的目标：  
一，我希望不计一切代价判断出少数类，得到最高的recall。  
二，我们希望追求最高的预测准确率，一切目的都是为了让accuracy更高，我们不在意recall或者AUC。  
三，我们希望达到recall，ROC和accuracy之间的平衡，不追求任何一个也不牺牲任何一个。  

#### 调recall（希望模型更多的捕获少数类）

```python
### 如果我们想更多的捕捉少数类，可以试着调clsss_weight,使得模型向着少数类的方向靠近
t0 = time()
for kernel in ["linear", "rbf", "sigmoid", "poly"]:
    svc = SVC(kernel=kernel
              , degree=1
              , gamma="auto"
              , class_weight="balanced"
              , cache_size=5000
             ).fit(Xtrain, Ytrain)
    result = svc.predict(Xtest)
    # recall召回率　捕捉少数类的能力
    score = svc.score(Xtest, Ytest)
    recall = recall_score(Ytest, result)
    auc = roc_auc_score(Ytest, svc.decision_function(Xtest))
    print("The kernle is {}, score : {}, recall : {}, auc : {}".format(kernel, score, recall, auc))
    print(datetime.datetime.fromtimestamp(time() - t0).strftime("%M:%S:%f"))
```

```python
## 确定模型更加适合线性核函数，无论从时间和性能上
## 接着调节class_weight 使得模型向着少数类的方向发展
t0 = time()
svc = SVC(kernel="linear"
          , degree=1
          , gamma="auto"
          , class_weight = {1:10} # 少数类的权重为10(正类)
          , cache_size=5000).fit(Xtrain, Ytrain)
result = svc.predict(Xtest)
# recall召回率　捕捉少数类的能力
score = svc.score(Xtest, Ytest)
recall = recall_score(Ytest, result)
auc = roc_auc_score(Ytest, svc.decision_function(Xtest))
print("The kernle is {}, score : {}, recall : {}, auc : {}".format(kernel, score, recall, auc))
print(datetime.datetime.fromtimestamp(time() - t0).strftime("%M:%S:%f"))
```

**结论**：虽然现在模型的准确率上去了，意味着我们的模型对于预测明天下雨的概率增大，而伴随着的是准确率的减低.

```python
# 此时的尝试在于是否存在一个使得模型稍微向少数样本的方向进行，却又能够提升准确率．
irange = np.linspace(0.01, 0.05, 10)
for i in irange:
    t0 = time()
    svc = SVC(kernel="linear"
              , gamma="auto"
              , class_weight = {1:1+i} # 少数类的权重为10(正类)
              , cache_size=5000).fit(Xtrain, Ytrain)
    result = svc.predict(Xtest)
    # recall召回率　捕捉少数类的能力
    score = svc.score(Xtest, Ytest)
    recall = recall_score(Ytest, result)
    auc = roc_auc_score(Ytest, svc.decision_function(Xtest))
    print("under ratio {}, score : {}, recall : {}, auc : {}".format(1+i, score, recall, auc))
    print(datetime.datetime.fromtimestamp(time() - t0).strftime("%M:%S:%f"))
```

#### 7.2 追求平衡

```python
# 调节C值, 惩罚项
score_all = []
recall_all = []
auc_all = []


crange = np.linspace(0.01, 20, 20)
for c in crange:
    t0 = time()
    svc = SVC(kernel="linear"
              , C=c
              , gamma="auto"
              , class_weight = "balanced" 
              , cache_size=5000).fit(Xtrain, Ytrain)
    result = svc.predict(Xtest)
    # recall召回率　捕捉少数类的能力
    score = svc.score(Xtest, Ytest)
    score_all.append(score)
    
    recall = recall_score(Ytest, result)
    recall_all.append(recall)
    
    auc = roc_auc_score(Ytest, svc.decision_function(Xtest))
    auc_all.append(auc)

    print("C : {}, score : {}, recall : {}, auc : {}".format(c, score, recall, auc))
    print(datetime.datetime.fromtimestamp(time() - t0).strftime("%M:%S:%f"))
    
plt.figure(figsize=(12, 6))
plt.plot(crange, score_all, label="accuracy")
plt.plot(crange, recall_all, label="recall")
plt.plot(crange, auc_all, label="auc")
plt.legend()
plt.show()
```

```python
t0 = time()
svc = SVC(kernel="linear"
          , C=1.0621052631578947
          , gamma="auto"
          , class_weight = "balanced" 
          , cache_size=5000).fit(Xtrain, Ytrain)
result = svc.predict(Xtest)
# recall召回率　捕捉少数类的能力
score = svc.score(Xtest, Ytest)
score_all.append(score)

recall = recall_score(Ytest, result)
recall_all.append(recall)

auc = roc_auc_score(Ytest, svc.decision_function(Xtest))
auc_all.append(auc)

print("score : {}, recall : {}, auc : {}".format(score, recall, auc))
print(datetime.datetime.fromtimestamp(time() - t0).strftime("%M:%S:%f"))
```
