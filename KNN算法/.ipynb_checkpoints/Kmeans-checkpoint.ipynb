```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from matplotlib import cm
```

```python
# 制造数据集
X, y = make_blobs(n_samples=500, n_features=2, centers=4, random_state=1)
```

```python
X.shape
```

```python
y.shape
```

```python
set(y)
```

```python
# 两个维度数据可视化
fig, ax = plt.subplots(1, figsize=(12, 6))

ax.scatter(X[:, 0]
           , X[:,1]
           , marker = "o"
           , s = 8
          )
plt.show()
```

```python
# 两类数据特征的分布
colors = ["red", "pink", "blue", "yellow"]
fig, ax = plt.subplots(1, figsize = (12, 6))
for i in range(4):
    ax.scatter(X[y == i, 0], X[y == i,  1]
               , c=colors[i]
               , marker="o"
               , s = 8)
plt.show()
```

```python
# 使用kmeans聚类算法计算总共有几类
n_clusters = 3
cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

```python
# 簇的质心
cluster.cluster_centers_
```

```python
# 所有簇的簇内平方和的总和
# total_inertia.append(cluster.inertia_)
```

```python
y_pred = cluster.labels_
pre = cluster.fit_predict(X)

# 两者结果相等
set(pre == y_pred)
```

```python
total_inertia = []

for i in range(3, 50):
    cluster = KMeans(n_clusters=i, random_state=0).fit(X)
    total_inertia.append(cluster.inertia_)

# 簇的个数越大,CSS越小, 评估指标无法进行定位
plt.figure(figsize=(12, 6))
plt.plot(range(3, 50), total_inertia)
plt.xlabel("n_clusters")
plt.ylabel("total_inertia")
plt.show()
```

```python
# 两类数据特征的分布
n_clusters = 3
cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
center = cluster.cluster_centers_
colors = ["red", "pink", "blue", "yellow"]
fig, ax = plt.subplots(1, figsize = (12, 6))

# 画出数据分布
for i in range(n_clusters):
    ax.scatter(X[cluster.labels_ == i, 0], X[cluster.labels_ == i,  1]
               , c=colors[i]
               , marker="o"
               , s = 8
              )

#　画出簇的质心的位置
ax.scatter(center[:, 0], center[:, 1]
           , marker="x"
           , s = 18
           , c="black"
          )
plt.title("the cluster in different n_clusters")
plt.show()
```

```python
# 基于轮廓系数的指标
from sklearn.metrics import silhouette_score
from time import time
t0 = time()

score = []

for i in range(3, 10):
    cluster = KMeans(n_clusters=i, random_state=0).fit(X)
    score.append(silhouette_score(X, cluster.labels_))

# 簇的个数越大,CSS越小, 评估指标无法进行定位
plt.figure(figsize=(12, 6))
plt.plot(range(3, 10), score)
plt.title("silhouette_score")
plt.xlabel("n_clusters")
plt.ylabel("total_inertia")
plt.show()
time()-t0
```

```python
from sklearn.metrics import silhouette_samples

#　计算每一个样本的轮廓系数
cluster = KMeans(n_clusters=4, random_state=0).fit(X)
silhouette_samples(X, cluster.labels_).mean()
```

```python
# 基于卡林斯基-哈拉巴斯指数的
from sklearn.metrics import calinski_harabasz_score
from time import time
t0 = time()
score = []

for i in range(3, 10):
    cluster = KMeans(n_clusters=i, random_state=0).fit(X)
    score.append(calinski_harabasz_score(X, cluster.labels_))

# 簇的个数越大,CSS越小, 评估指标无法进行定位
plt.figure(figsize=(12, 6))
plt.plot(range(3, 10), score)
plt.title("silhouette_score")
plt.xlabel("n_clusters")
plt.ylabel("total_inertia")
plt.show()
time()-t0
```

### 案例：基于轮廓系数来调整n_clusters

```python
for n_clusters in range(2, 7):

    fig ,(ax1, ax2) = plt.subplots(1, 2)
    # ax1绘制不同的簇类
    fig.set_size_inches(19, 6)
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, X.shape[0] + (n_clusters + 1)*50])
    cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
    silhouette_avg = silhouette_score(X, cluster.labels_)
    cluster_label = cluster.labels_
    silhouette_samples_values = silhouette_samples(X, cluster_label)
    y_lower = 10

    for i in range(n_clusters):
        # 计算每一一个簇的轮廓系数
        ith_cluster_silhouette_values = silhouette_samples_values[cluster_label == i]
        ith_cluster_silhouette_values.sort()
        # 簇内的个数
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = cm.nipy_spectral(float(i)/n_clusters)

        ax1.fill_betweenx(np.arange(y_lower, y_upper)
                          , ith_cluster_silhouette_values
                          , facecolor= color
                          , alpha=0.7)
       # 设置簇内标签
        ax1.text(-0.05
                 , y_lower + 0.5*size_cluster_i
                 , str(i)
                )
        y_lower = y_upper + 10

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    ax1.axvline(x = silhouette_avg, color = "red", linestyle = "--")
    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 可视化簇的质心
    center = cluster.cluster_centers_

    color = cm.nipy_spectral(cluster_label.astype(float)/n_clusters)
    # 数据划分
    ax2.scatter(X[:, 0], X[:, 1]
                , marker="p"
                , s=8
                , c=color
               )
    # 绘制质心
    ax2.scatter(center[:, 0], center[:, 1]
                , marker="X"
                , s=100
                , c="red"
                ,alpha=1
               )
    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
    "with n_clusters = %d" % n_clusters)
                 , fontsize=14
                 , fontweight='bold'
                )
    print("\n")
    plt.show()
```

### 让迭代停下来 

**目的**:在完全收敛之前,我们也可以使用max_iter和tol让迭代停下来

```python
random = KMeans(n_clusters = 10,init="random",max_iter=10,random_state=420).fit(X)
y_pred_max10 = random.labels_
silhouette_score(X,y_pred_max10)
```

```python
random = KMeans(n_clusters = 10,init="random",max_iter=20,random_state=420).fit(X)
y_pred_max20 = random.labels_
silhouette_score(X,y_pred_max20)
```
