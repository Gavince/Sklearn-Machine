```python
import matplotlib.pylab as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
rnd = np.random.RandomState(42)
X = rnd.uniform(-3, 3, size=100) 
y = np.sin(X) + rnd.normal(size=len(X)) / 3 
```

```python
plt.scatter(X, y, marker="o")
plt.show()
```

```python
X = X.reshape(-1, 1)
```

```python
X.shape
```

```python
DEGREE = 6
LR = LinearRegression().fit(X, y)
tree = DecisionTreeRegressor(random_state=1).fit(X, y)
# 测试集
LR_n = LinearRegression().fit(PolynomialFeatures(degree=DEGREE, include_bias=False).fit_transform(X), y)
line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1, 1)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(line, LR.predict(line), c = "b", label="LR")
ax.plot(line, tree.predict(line), c = "g", label="tree")
ax.scatter(X[:, 0], y, c="r", marker="o", label="original")
plt.legend(loc=4)
plt.tight_layout()
plt.show()
```

结论：树模型天生过拟合,更加容易拟合数据，而线性模型天生对于线性数据有较好的依赖性，对于非线性的数据，其效果一般不好，也只能模拟出数据的大概走向。

### 分箱

```python
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
```

```python
# 制作哑变量参数
enc = KBinsDiscretizer(n_bins=20, encode="onehot")
X_bined = enc.fit_transform(X)
X_bined
```

```python
pd.DataFrame(X_bined.toarray()).head()
```

```python
# 箱体边界
enc.bin_edges_
```

```python
# 直接使用LR进行训练
linear = LinearRegression().fit(X_bined, y)
```

```python
## 需要将训练数据进行转化
line_enc = enc.transform(line)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6), sharey=True)
ax1.scatter(line, LR.predict(line), c = "b", label="LR")
ax1.plot(line, tree.predict(line), c = "g", label="tree")
ax1.scatter(X[:, 0], y, c="r", marker="o", label="original")
ax1.set_xlabel("Inputer feature")
ax1.set_ylabel("Regression output")
ax1.set_title("The result before discretization")
ax1.legend(loc=4)

ax2.scatter(line, linear.predict(line_enc), c = "b", label="LR")
ax2.plot(line, tree.predict(line), c = "g", label="tree")
ax2.scatter(X[:, 0], y, c="r", marker="o", label="original")
ax2.set_xlabel("Inputer feature")
ax2.set_title("The result after discretization")
ax2.vlines(enc.bin_edges_[0], *plt.gca().get_ylim(), linewidth=1, alpha=0.2)
ax2.legend(loc=4)

# plt.legend(loc=4)
plt.tight_layout()
plt.show()
```

结论：箱体分的越小，则预测结果越容易过拟合，应选取适当的分箱个数。

#### 选取合适的分箱

```python
from sklearn.model_selection import cross_val_score as CVS
import numpy as np
```

```python
pred,score,var = [], [], []
binsrange = [2,5,10,15,20,30]
for i in binsrange:
    #实例化分箱类
    enc = KBinsDiscretizer(n_bins=i,encode="onehot")
    #转换数据
    X_binned = enc.fit_transform(X)
    line_binned = enc.transform(line)
    #建立模型
    LinearR_ = LinearRegression()
    #全数据集上的交叉验证
    cvresult = CVS(LinearR_,X_binned,y,cv=5)
    score.append(cvresult.mean())
    var.append(cvresult.var())
    #测试数据集上的打分结果
    pred.append(LinearR_.fit(X_binned,y).score(line_binned,np.sin(line)))
    #绘制图像
plt.figure(figsize=(6,5))
plt.plot(binsrange,pred,c="orange",label="test")
plt.plot(binsrange,score,c="k",label="full data")
## 验证方差和准确度
plt.plot(binsrange,score+np.array(var)*0.5,c="red",linestyle="--",label = "var")
plt.plot(binsrange,score-np.array(var)*0.5,c="red",linestyle="--")
plt.legend()
plt.show()
```

### 多项式

```python
from sklearn.preprocessing import PolynomialFeatures
# 三步走原则
```

```python

X1 = np.arange(1, 4).reshape(-1, 1)
X1
```

```python
poly = PolynomialFeatures(5)
poly_x = poly.fit_transform(X1)

# 系数分别为：w0(偏差), (w1, ......Wn:特征维度)
poly_x
```

```python
# 特征树不包含偏差项
poly = PolynomialFeatures(5, include_bias=False, interaction_only=True)
poly_x = poly.fit_transform(X1)

# 系数分别为：w0(偏差), (w1, ......Wn:特征维度)
poly_x
```

```python
# 特征树不包含偏差项
poly = PolynomialFeatures(5, include_bias=False, interaction_only=False)
poly_x = poly.fit_transform(X1)

# 系数分别为：w0(偏差), (w1, ......Wn:特征维度)
poly_x
```

```python
Lr = LinearRegression()
Lr.fit(poly_x, rnd.randn(3))
```

```python
Lr.intercept_
```

```python
Lr.coef_
```

```python
# 多维数据表示(随着最高次数的额上升，维度数也逐渐增加)
X_n = np.arange(6).reshape(3, 2)
poly_n = PolynomialFeatures(100, include_bias=False)
poly_n = poly_n.fit_transform(X_n)
poly_n.shape
```

```python
# 消除多重共线性，只保留交差项
X_n = np.arange(6).reshape(3, 2)
poly_n = PolynomialFeatures(2, include_bias=False, interaction_only=True)
poly_n = poly_n.fit_transform(X_n)
poly_n
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(line, LR.predict(line), c = "b", label="LR")
ax.scatter(line, LR_n.predict(PolynomialFeatures(degree=DEGREE,include_bias=False).fit_transform(line))
           , c = "pink", label="LR_n")
ax.plot(line, tree.predict(line), c = "g", label="tree")
ax.scatter(X[:, 0], y, c="r", marker="o", label="original")
plt.legend(loc=4)
plt.tight_layout()
plt.show()
```

结论：使用多项式后使得原本不能够线性数据可以很好的拟合非线性数据（升高维度）

```python
from sklearn.datasets import fetch_california_housing as fch
import pandas as pd
```

```python
housevalue = fch()
X = pd.DataFrame(housevalue.data)
y = housevalue.target
housevalue.feature_names
```

```python
X.columns = ["住户收入中位数","房屋使用年代中位数","平均房间数目"
            ,"平均卧室数目","街区人口","平均入住率","街区的纬度"
             ,"街区的经度"]

poly = PolynomialFeatures(degree=2).fit(X,y)
```

```python
poly.get_feature_names(X.columns)
```

```python
X_ = poly.transform(X)
```

```python
reg = LinearRegression().fit(X_, y)
reg
```

```python
df = pd.DataFrame([poly.get_feature_names(X.columns), reg.coef_.tolist()]).T
```

```python
df.columns = ["feature", "coef"]
```

```python
# 多项式回归可以很好的的解释性参数，可以很好地通过可视化来判
# 断组合后的特征那一个更具有更多的关注度
df.sort_values(by="coef")
```

```python
poly = PolynomialFeatures(degree=4).fit(X,y)
X_ = poly.transform(X)
reg = LinearRegression().fit(X,y)
reg.score(X,y)
```

```python
from time import time
time0 = time()
reg_ = LinearRegression().fit(X_,y)
print("R2:{}".format(reg_.score(X_,y)))
print("time:{}".format(time()-time0))
```

```python
#假设使用其他模型？
from sklearn.ensemble import RandomForestRegressor as RFR
time0 = time()
print("R2:{}".format(RFR(n_estimators=100).fit(X,y).score(X,y)))
print("time:{}".format(time()-time0))
```

结论：尽管多项式回归，具有较好的R2，但是在计算的时间的上是线性回归的十倍左右，时间复杂度增加，所以，可以根据自己的需要选择合适的模型来追求时间和空间复杂度的平衡。
